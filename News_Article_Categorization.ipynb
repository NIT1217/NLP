#News Article Categorization
1. Scenario:You run a media platform and want to tag incoming articles into categories like Politics, Sports, Tech, etc.
2. Task: Train a classifier that uses average sentenceTransformer embeddings of each article as input features for classification.

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
data = {
    'text': [
        "The prime minister announced a new economic policy today.",
        "The football team won their third consecutive championship.",
        "Apple released its latest iPhone with improved camera features.",
        "The movie won several awards at the international film festival.",
        "The parliament debated the new environmental bill."
    ],
    'label': ['Politics', 'Sports', 'Tech', 'Entertainment', 'Politics']
}
```

```python
embedded_text  = model.encode(data["text"])
embedded_label = model.encode(data["label"])
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(embedded_text,data["label"])
```

```python
new_news = "The government introduced a new tax policy."
embeded_news = model.encode(new_news).reshape(1,-1)
print(clf.predict(embeded_news)[0])
```
